```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def option_payout(S, K, typ="call",prem=0.0):
    if typ == "call":
        typ = "Call"
        p = 1
    elif typ == "put":
        typ == "Put"
        p = -1    
    payout = np.maximum(p*(S - K), 0) - prem
    return payout, typ
```

```python
S = 120
K = 110
prem = 5.4
typ="put"
```

```python
n = 50
spot_prices = np.linspace(0.8*S, 1.2*S, n)
payouts = np.zeros(n)
```

```python
for i in range(len(spot_prices)): 
    payouts[i], typ = option_payout(S=spot_prices[i], K=K,prem=prem, typ="call")
```

```python
plt.figure(figsize=(7,4))
plt.plot(spot_prices, payouts)
plt.axvline(x = K, color="red", linestyle = '--', label=f"Strike Price")
plt.axhline(y = 0, color="green", linestyle = '--', label="Breakeven")
plt.title(f"{typ} Option Payoff")
plt.xlabel("Spot Prices")
plt.ylabel("Payout")
plt.legend()
plt.show()
```
